```python
def count_num_price_changed(df):

    num_price_changed = 0
    last_price = df["last"][0]
    for i in range(len(df)):
        if df["last"][i] != last_price:
            num_price_changed += 1
            last_price = df["last"][i]
            
    return num_price_changed
        
```
